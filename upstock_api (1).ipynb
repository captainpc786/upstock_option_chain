```python
def getdata(symbol, intervel , date_from ,date_to ):

    inst_key = getinstrumentkey(symbol)
    url = f'https://api.upstox.com/v3/historical-candle/{inst_key.values[0]}/{intervel}/{date_to}/{date_from}'

    headers = {
         'Accept': 'application/json'
         }

    response = requests.get(url, headers=headers)

    # Check the response status
    if response.status_code == 200:
        # Do something with the response data (e.g., print it)
        print(response.json())
    else:
        # Print an error message if the request was not successful
        print(f"Error: {response.status_code} - {response.text}")

    data = pd.DataFrame.from_dict(response.json()['data']['candles'])
    cols = ['datetime','open','high','low','close','volume','open interest']
    data.columns = cols
    data['datetime'] = data.apply(convert_date, axis=1)
    data['datetime'] = pd.to_datetime(data['datetime'])
    return data
```

```python
def getdata(symbol, interval, date_from, date_to):
    inst_key = getinstrumentkey(symbol)

    url = f'https://api.upstox.com/v3/historical-candle/{inst_key.values[0]}/{interval}/{date_to}/{date_from}'
    headers = {
        'Accept': 'application/json'
    }

    response = requests.get(url, headers=headers)

    if response.status_code != 200:
        print(f"Error: {response.status_code} - {response.text}")
        return None

    response_data = response.json()

    # Safeguard: check for valid candle data
    if 'data' not in response_data or 'candles' not in response_data['data'] or not response_data['data']['candles']:
        print("No candle data available.")
        return pd.DataFrame()  # return empty DataFrame safely

    data = pd.DataFrame.from_dict(response_data['data']['candles'])

    # Only rename columns if data is not empty
    if not data.empty:
        cols = ['datetime', 'open', 'high', 'low', 'close', 'volume', 'open interest']
        data.columns = cols
        data['datetime'] = pd.to_datetime(data['datetime'])

    return data
```
